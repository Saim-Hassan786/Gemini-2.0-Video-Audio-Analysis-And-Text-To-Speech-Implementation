```python
import time
Intro_video = "/content/Introduction.mp4"

def upload_video(video_file_name):
  video_file = client.files.upload(path = video_file_name)
  while video_file.state=="PROCESSING":
    print("Video Is Being Processed, Kindly Wait!")
    time.sleep(10)
    video_file = client.files.get(name=video_file.name or "")
  if video_file.state =="SUCCESS":
    pass
  elif video_file.state=="FAILED":
    raise ValueError(video_file.state)
  print(f'Video processing complete: ' + (video_file.uri or ""))
  return video_file

video_intro = upload_video(Intro_video)
```

**The following code helps manage audio recording and processes data asynchronously. First, it imports necessary modules: `contextlib` to manage context (like opening and closing files), `wave` to work with `.wav` audio files, and `IPython.display` for audio playback.**

**The `async_enumerate` function is an asynchronous function. It takes an iterable (`it`) and asynchronously loops through it, yielding each item along with its index (`n`). This allows for efficient, non-blocking iteration, which is useful when dealing with tasks that take time, like processing audio data.**

**Next, the `audio_recording` function is a context manager, defined using `@contextlib.contextmanager`. A context manager ensures that resources, like files, are properly opened and closed. In this case, it opens a `.wav` audio file with specific properties, such as the sample width (size of each audio sample), number of channels (1 for mono, 2 for stereo), and the sample rate (the number of samples per second, 24000 in this case). Inside the context manager, the `wave.open` function creates or overwrites the file, and the `yield wf` statement temporarily returns the `wave` file object, allowing for audio data to be written into it. Once the block of code that uses the `audio_recording` function finishes, the file is automatically closed.**

**Overall, the code sets up a framework for recording and processing audio in a `.wav` format asynchronously, making it suitable for audio data to be handled efficiently.**

```python
import contextlib
import wave
from IPython.display import display, Audio

async def async_enumerate(it):
  n = 0
  async for item in it:
    yield n, item
    n +=1

@contextlib.contextmanager
def audio_recording(filename,channels=1, rate = 24000, sample_width = 2):
  with wave.open(filename, "wb") as wf:
     wf.setsampwidth(sample_width)
     wf.setnchannels(channels)
     wf.setframerate(rate)
     yield wf
```

**Now, The key `"generation_config"` holds another dictionary that contains configuration details.
```python
config = {"generation_config":{"response_modalities":["AUDIO"]}}
```

**This code sets up an asynchronous process to send a message, receive an audio response, and save the response to a `.wav` file. It starts by connecting to an AI model using `client.aio.live.connect`, which allows the program to interact with the model in real-time. The `config` variable specifies that the response should be in audio format, and `MODEL` refers to the specific AI model being used.**

**Once the connection is made, it defines a file called `greeting.wav` to store the audio response. The `audio_recording(filename)` context manager is used to handle the audio file, which ensures the file is opened for writing and will automatically close when done. Inside the context manager, the message `"Hi there myself Saim Hassan Akhtar, tell me something about Agentic AI"` is sent to the AI model using `session.send(msg, end_of_turn=True)`. This method sends the message and signals the end of the user's input.**

**Next, the program waits for the AI's response using `session.receive()`, which retrieves the response from the model. The response is processed asynchronously using the `async_enumerate` function, which allows for efficient, non-blocking iteration over the response data. As each chunk of response data is received, it is written into the `greeting.wav` file using `wf.writeframes(response_chunks.data)`.**

**If there is data in the response (`if response_chunks.data is not None`), it gets written into the file. The `n == 0` check ensures that a message is printed when the first chunk of audio data is received, indicating that the audio response is being generated. After all the data is written, the audio file is ready, and it is played using `display(Audio(filename, autoplay=True))`, which automatically plays the generated audio response.**

```python
async with client.aio.live.connect(model=MODEL,config=config) as session:
  filename = "greeting.wav"
  with audio_recording(filename)as wf:
    msg = "Hi there myself Saim Hassan Akhtar, tell me about Agentic Artificial Intelligence "
    print(msg)
    await session.send(msg,end_of_turn= True)
    response = session.receive()

    async for n,response_chunks in async_enumerate(response):
      if response_chunks.data is not None:
        wf.writeframes(response_chunks.data)
        if n==0:
          print(".......Your Audio Response Is Generating......")

display(Audio(filename,autoplay=True))
```

**Now , I customized the above above to give the response to the user by taking prompt from it , this code defines a function `generate_audio_response` that sends a text message to an AI model, receives an audio response, and saves the audio to a `.wav` file. If the first chunk is received, a message ".......Your Audio Response Is Generating......" is printed to indicate progress.**

**Once all data is written to the audio file, the function plays the audio file automatically using `display(Audio(filename, autoplay=True))`. This ensures that the user can hear the AI's response immediately after it's generated. The whole process is asynchronous, meaning it doesn’t block other tasks while waiting for the response.**
